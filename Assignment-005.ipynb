# Testing Function

This function is needed to test your code. Be sure to run the cell below before you try to test your code.

````
def check_output(actual_output,desired_output,identifier=''):
    identifier = str(identifier)
    if len(identifier) > 0:
        print('Test ID:',identifier)

    if actual_output == desired_output:
        print('Success')
    else:
        print('Failure')
        print('Desired output:',desired_output)
        print('Actual output:',actual_output)

    print('')
````

#Problem 3

**Implement the following helper methods in your matrix class.**

* `get_pivot_row(self, column_index)`: returns the index of the topmost row that has a nonzero entry in the desired `column_index` and such that all entries left of `column_index` are zero. Otherwise, if no row exists, return `None`.

* `swap_rows(self, row_index1, row_index2)`: swap the row at `row_index1` with the row at `row_index2`.

* `scale_row(self, row_index)`: divide the entire row at `row_index` by the row's first nonzero entry.

* `clear_below(self, row_index)`:

 * Let $j$ be the column index of the first nonzero entry in the row at `row_index`.
 * Subtract multiples of the row at `row_index` from the rows below, so that for any row below `row_index`, the entry at column $j$ is zero.

* `clear_above(self, row_index)`:

 * Let $j$ be the column index of the first nonzero entry in the row at `row_index`.
 * Subtract multiples of the row at `row_index` from the rows above, so that for any row above `row_index`, the entry at column $j$ is zero.

**Watch out!**

* Remember that the first row/column of a matrix has the index `0`, not `1`.

* If `row1` is "below" `row2` in a matrix, then `row1` actually has a higher index than `row2`. This is because the `0` index corresponds to the very top row.

**Example:**

```
>>> A = Matrix(elements = [[0, 1, 2],
                           [3, 6, 9],
                           [2, 6, 8]])
>>> A.get_pivot_row(0)          (TEST ID: 1)
1
>>> A.swap_rows(0,1)
>>> A.show()          (TEST ID: 2)
[3, 6, 9]
[0, 1, 2]
[2, 6, 8]
>>> A.scale_row(0)
>>> A.show()          (TEST ID: 3)
[1, 2, 3]
[0, 1, 2]
[2, 6, 8]
>>> A.clear_below(0)
>>> A.show()          (TEST ID: 4)
[1, 2, 3]
[0, 1, 2]
[0, 2, 2]
>>> A.get_pivot_row(1)          (TEST ID: 5)
1
>>> A.scale_row(1)
>>> A.show()          (TEST ID: 6)
[1, 2, 3]
[0, 1, 2]
[0, 2, 2]
>>> A.clear_below(1)
>>> A.show()          (TEST ID: 7)
[1, 2, 3]
[0, 1, 2]
[0, 0, -2]
>>> A.get_pivot_row(2)          (TEST ID: 8)
2
>>> A.scale_row(2)
>>> A.show()          (TEST ID: 9)
[1, 2, 3]
[0, 1, 2]
[0, 0, 1]
>>> A.clear_above(2)
>>> A.show()          (TEST ID: 10)
[1, 2, 0]
[0, 1, 0]
[0, 0, 1]
>>> A.clear_above(1)
>>> A.show()          (TEST ID: 11)
[1, 0, 0]
[0, 1, 0]
[0, 0, 1]
```

````
A = Matrix(elements = [[0, 1, 2],
                       [3, 6, 9],
                       [2, 6, 8]])
check_output(A.get_pivot_row(0),1,1)
A.swap_rows(0,1)
check_output(A.elements,[[3, 6, 9],[0, 1, 2],[2, 6, 8]],2)
A.scale_row(0)
check_output(A.elements,[[1, 2, 3],[0, 1, 2],[2, 6, 8]],3)
A.clear_below(0)
check_output(A.elements,[[1, 2, 3],[0, 1, 2],[0, 2, 2]],4)
check_output(A.get_pivot_row(1),1)
A.scale_row(1)
check_output(A.elements,[[1, 2, 3],[0, 1, 2],[0, 2, 2]],6)
A.clear_below(1)
check_output(A.elements,[[1, 2, 3],[0, 1, 2],[0, 0, -2]],7)
check_output(A.get_pivot_row(2),8)
A.scale_row(2)
check_output(A.elements,[[1, 2, 3],[0, 1, 2],[0, 0, 1]],9)
A.clear_above(2)
check_output(A.elements,[[1, 2, 0],[0, 1, 0],[0, 0, 1]],10)
A.clear_above(1)
check_output(A.elements,[[1, 0, 0],[0, 1, 0],[0, 0, 1]],11)
````
