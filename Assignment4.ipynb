# Testing Function

This function is needed to test your code. Be sure to run the cell below before you try to test your code.

````
def checkOutput(actualOutput,desiredOutput,identifier=''):
  identifier = str(identifier)
  if len(identifier) > 0:
    print('Test ID:',identifier)

  if actualOutput == desiredOutput:
    print('Success')
  else:
    print('Failure')
    print('Desired output:',desiredOutput)
    print('Actual output:',actualOutput)

  print('')
````
